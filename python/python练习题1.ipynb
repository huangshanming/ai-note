### （一）四个数字1、2、3、4组成互不相同且无重复的三位数

```python
data = [1, 2, 3, 4]
for i in data: 
    for j in data:
        for k in data:
            if i != j and i != k and j != k:
                print ('{a}{b}{c}'.format(a=i, b=j, c=k))
```

### （二）企业发放的奖金根据利润提成。利润 (I) 低于或等于 10 万元时，奖金可提 10%；利润高于 10 万元，低于 20 万元时，低于 10 万元的部分按 10% 提成，高于 10 万元的部分，可提成 7.5%；20 万到 40 万之间时，高于 20 万元的部分，可提成 5%；40 万到 60 万之间时高于 40 万元的部分，可提成 3%；60 万到 100 万之间时，高于 60 万元的部分，可提成 1.5%，高于 100 万元时，超过 100 万元的部分按 1% 提成，从键盘输入当月利润 I，求应发放奖金总数？

```python
l = input("请输入当月利润l")

def getBouns(lr, num ,radio, bouns):
    lr = float(lr)
    if lr > num:
        bouns += (lr - num) * radio
        lr = num
    return lr, bouns

bouns = 0
l, bouns = getBouns(l, 1000000, 0.01, bouns)
l, bouns = getBouns(l, 600000, 0.015, bouns)
l, bouns = getBouns(l, 400000, 0.03, bouns)
l, bouns = getBouns(l, 200000, 0.05, bouns)
l, bouns = getBouns(l, 100000, 0.075, bouns)
l, bouns = getBouns(l, 0, 0.1, bouns)
print("奖金总数为：{}".format(bouns))
```

```python
# 1. 获取用户输入的利润，并做输入校验（确保输入是合法数字）
while True:
    try:
        profit_input = input("请输入当月利润（单位：万元）：")
        # 将输入的字符串转为浮点数（支持整数/小数输入）
        profit = float(profit_input)
        if profit < 0:
            print("利润不能为负数，请重新输入！")
            continue
        break  # 输入合法则退出循环
    except ValueError:
        print("输入格式错误，请输入数字（如15、28.5）！")

# 2. 初始化奖金总额
bonus = 0.0

# 3. 分段计算奖金（从高到低判断，逻辑更清晰）
# 超过100万的部分：按1%提成
if profit > 100:
    bonus += (profit - 100) * 0.01
    profit = 100  # 剩余利润限定为100万，继续计算下一段
# 60万-100万的部分：按1.5%提成
if profit > 60:
    bonus += (profit - 60) * 0.015
    profit = 60
# 40万-60万的部分：按3%提成
if profit > 40:
    bonus += (profit - 40) * 0.03
    profit = 40
# 20万-40万的部分：按5%提成
if profit > 20:
    bonus += (profit - 20) * 0.05
    profit = 20
# 10万-20万的部分：按7.5%提成
if profit > 10:
    bonus += (profit - 10) * 0.075
    profit = 10
# 10万及以下的部分：按10%提成
bonus += profit * 0.1

# 4. 输出结果（保留2位小数，符合金额显示习惯）
print(f"当月应发放奖金总额：{bonus:.2f} 万元")
```
