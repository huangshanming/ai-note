### 14.广播机制与向量化（核心：广播规则、替代循环）
核心目标

练习题 1：广播规则验证
题目要求
验证以下维度组合是否满足广播规则，若满足则输出结果维度：
(3,4) 和 (4,) → 满足，结果 (3,4)
(2,3,4) 和 (3,4) → 满足，结果 (2,3,4)
(2,3) 和 (3,2) → 不满足
(5,1,6) 和 (4,6) → 满足，结果 (5,4,6)
手动实现广播：将数组arr1 = np.array([[1],[2],[3]])（3×1）与arr2 = np.array([4,5,6])（1×3）相加，验证结果维度为 3×3。


```python
def check_broadcast(shape1, shape2):
    try: 
        a = np.ones(shape1)
        b = np.ones(shape2)
        res = a + b
        return f"满足广播规则，结果纬度：{res.shape}"
    except ValueError as e:
        return f"不满足广播规则，错误：{e}"
    
cases = [
    ((3,4), (4,)),
    ((2,3,4), (3,4)),
    ((2,3), (3,2)),
    ((5,1,6), (4,6))
]

print("1. 广播规则验证:")
for i, (s1, s2) in enumerate(cases):
    print(f"案例{i+1}: {s1} 和 {s2} → {check_broadcast(s1, s2)}")
```

### 16.线性代数与矩阵运算（核心：矩阵乘法、逆矩阵、特征值）
核心目标
掌握 NumPy 线性代数模块（np.linalg）的常用功能，适配机器学习 / 数据分析中的矩阵运算场景。
练习题 1：基础矩阵运算
题目要求
创建两个矩阵A = np.array([[1,2],[3,4]])和B = np.array([[5,6],[7,8]])，计算：
矩阵乘法（A・B）；
矩阵的行列式（det (A)）；
矩阵 A 的逆矩阵（A⁻¹）；
验证 A・A⁻¹ = 单位矩阵（忽略浮点误差）；
解线性方程组：
{ 
2x+3y=8
5x−y=3
​

```python
import numpy as np

# 1. 矩阵运算
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

# 矩阵乘法（注意：*是元素级，@是矩阵乘法）
mat_mul = A @ B  # 等价于np.matmul(A, B)
print(A)
print(B)
print("1. 矩阵乘法 A·B:\n", mat_mul)

# 行列式
det_A = np.linalg.det(A)
print("\n矩阵A的行列式:", det_A)

# 逆矩阵
inv_A = np.linalg.inv(A)
print("\n矩阵A的逆矩阵:\n", inv_A)

# 验证A·A⁻¹ = 单位矩阵
identity = A @ inv_A
print("\nA·A⁻¹（近似单位矩阵）:\n", np.round(identity, 0))  # 四舍五入消除浮点误差

# 2. 解线性方程组 Ax = b
# 方程组：2x+3y=8, 5x-y=3 → A=[[2,3],[5,-1]], b=[8,3]
A_eq = np.array([[2,3],[5,-1]])
b_eq = np.array([8,3])
x = np.linalg.solve(A_eq, b_eq)
print("\n2. 线性方程组解 (x,y):", x)
# 验证
print("验证: 2x+3y =", 2*x[0] + 3*x[1])
print("验证: 5x-y =", 5*x[0] - x[1])
```

### 17.特征值与矩阵分解
题目要求
计算对称矩阵arr = np.array([[4,2],[2,4]])的特征值和特征向量；
对矩阵arr = np.random.rand(3,3)进行奇异值分解（SVD）；
计算矩阵的秩、迹（对角线元素和）。

```python
import numpy as np

np.random.seed(42)

# 1. 特征值与特征向量
arr_sym = np.array([[4,2],[2,4]])
eigenvals, eigenvecs = np.linalg.eig(arr_sym)
print("1. 特征值:", eigenvals)
print("特征向量（每列是一个特征向量）:\n", eigenvecs)

# 验证：A·v = λ·v
v = eigenvecs[:, 0]  # 第一个特征向量
lambda_v = eigenvals[0]  # 第一个特征值
print("\n验证 A·v = λ·v:")
print("A·v =", arr_sym @ v)
print("λ·v =", lambda_v * v)

# 2. 奇异值分解（SVD）
arr_rand = np.random.rand(3,3)
U, S, VT = np.linalg.svd(arr_rand)
print("\n2. SVD分解结果:")
print("U矩阵（左奇异向量）:\n", np.round(U, 2))
print("奇异值S:", np.round(S, 2))
print("VT矩阵（右奇异向量转置）:\n", np.round(VT, 2))

# 3. 秩和迹
rank = np.linalg.matrix_rank(arr_rand)
trace = np.trace(arr_rand)
print("\n3. 矩阵秩:", rank)
print("矩阵迹:", trace)
```

### 18.高级技巧与性能优化（核心：花式索引、ufunc、内存优化）
核心目标
掌握 NumPy 的高级用法，解决实际场景中的性能和内存问题。
练习题 1：花式索引与条件筛选进阶
题目要求
从数组arr = np.random.randint(0, 100, (10, 10))中：
取出所有偶数行、奇数列的元素；
取出前 5 行中大于 50 的元素，并按降序排序；
将数组中第 3 行的元素替换为该行元素的均值；
用np.where()实现：将数组中小于 30 的元素设为 0，大于等于 30 的设为原值。

```python
import numpy as np

np.random.seed(42)
arr = np.random.randint(0, 100, (10, 10))
print("原始数组:\n", arr)

# 1. 花式索引
# 偶数行（0,2,4,6,8）、奇数列（1,3,5,7,9）
even_rows_odd_cols = arr[::2, 1::2]  # ::2是步长2（偶数行），1::2是从1开始步长2（奇数列）
print("\n1. 偶数行、奇数列元素:\n", even_rows_odd_cols)

# 前5行中大于50的元素，降序排序
top5_gt50 = arr[:5][arr[:5] > 50]
top5_gt50_sorted = np.sort(top5_gt50)[::-1]  # 排序后反转（降序）
print("\n前5行大于50的元素（降序）:", top5_gt50_sorted)

# 替换第3行为该行均值
arr[3] = arr[3].mean()
print("\n替换第3行后的数组:\n", arr)

# 2. np.where()条件替换
arr_where = np.where(arr < 30, 0, arr)
print("\n2. np.where替换后（<30→0）:\n", arr_where)
```

### 19.性能优化与内存管理
题目要求
对比np.ufunc（通用函数）与 Python 循环的性能：计算数组arr = np.arange(1e6)的正弦值；
用np.memmap创建内存映射数组（处理大文件，避免加载全部数据到内存）；
将数组arr = np.random.rand(1000, 1000)转换为更节省内存的float32类型，计算内存占用变化。

```python
import numpy as np
import time
import os

# 1. ufunc vs 循环（计算正弦值）
arr = np.arange(1000000, dtype=np.float64)

# Python循环
start = time.time()
sin_loop = np.zeros_like(arr)
for i in range(len(arr)):
    sin_loop[i] = np.sin(arr[i])
time_loop = time.time() - start

# NumPy ufunc（向量化）
start = time.time()
sin_ufunc = np.sin(arr)
time_ufunc = time.time() - start

print(f"1. 循环计算正弦值时间: {time_loop:.6f} 秒")
print(f"   ufunc计算正弦值时间: {time_ufunc:.6f} 秒")
print(f"   性能提升: {time_loop/time_ufunc:.2f} 倍")

# 2. 内存映射数组（处理大文件）
# 创建大数组并保存到文件
big_arr = np.random.rand(10000, 10000).astype(np.float32)
np.save("big_arr.npy", big_arr)
print(f"\n2. 大数组文件大小: {os.path.getsize('big_arr.npy')/1024/1024:.2f} MB")

# 用memmap加载（仅映射，不加载全部数据）
memmap_arr = np.load("big_arr.npy", mmap_mode="r")
print("内存映射数组形状:", memmap_arr.shape)
print("内存映射数组前2×2元素:\n", memmap_arr[:2, :2])

# 3. 内存优化（float64 → float32）
arr = np.random.rand(1000, 1000)  # 默认float64
mem_64 = arr.nbytes / 1024 / 1024  # 转MB
arr_32 = arr.astype(np.float32)
mem_32 = arr_32.nbytes / 1024 / 1024

print(f"\n3. float64内存占用: {mem_64:.2f} MB")
print(f"   float32内存占用: {mem_32:.2f} MB")
print(f"   内存节省比例: {1 - mem_32/mem_64:.2f}")

# 清理文件
os.remove("big_arr.npy")
```
